# COVID-19: A Breath of Fresh Air for Bangalore?
#### Looking at the Lockdown for Lessons on Bettering Air Quality.

The air we breathe is crucial to not only our health, but our experience of the world around us as well and while the novel coronavirus has infamously uprooted our idea of 'normal' and left millions gasping for breath, it has also given us a peek into the new normal and how one could take aspects of this new normal to make post-covid life, better. One such aspect being its impact on air quality/pollution.

In this notebook, we will use statistical methods to explore the impact of the virus, and more specifically the lockdowns on air quality and pollution in Bangalore.

#### Variables Explored:

1. Air Quality - AQI, Particulate Matter (2.5 and 10), NO2.
2. Urban Growth - Night-time lights data.
3. Urban Vegetation - Normalized Difference Vegetation Index.

#### Geographical Focus:

The geographical focus of this analysis will be on the City of Bangalore, primary in the urban center but also in the rural extensions of the city in some cases.

#### Data:

1. Air Quality Data - [[Kaggle]](https://www.kaggle.com/rohanrao/air-quality-data-in-india) - Used as the base variable in hypothesis testing, and correlation analysis with the other variables.
2. Night-time Lights Data - [Google Earth Engine] - Used to analyse urban growth, change in transportation trends, and in correlation analysis with Air Quality and more specifically, particulate matter.
3. Sentinel-2 L2A Satellite Images - [[Sentinel Hub]](https://apps.sentinel-hub.com/eo-browser/) - Used to analyse change in urban vegetation and its impact on Air Quality and more specifically, Nitrogen Dioxide.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
import scipy.stats as st
import datetime as dt
import os
from copy import deepcopy
```

```python
df_aqi_full = pd.read_csv('data/city_day.csv')
df_aqi = df_aqi_full[(df_aqi_full['City']=='Bengaluru') & (df_aqi_full.AQI.notnull())].reset_index(drop=True)
```

### Exploratory Data Analysis:

###### Univariate Analysis:

```python
def aqi_clean(df_aqi):
    df_aqi['Date'] = df_aqi['Date'].apply(lambda x: pd.to_datetime(x))
    return df_aqi.set_index('Date', drop=True)

df2 = aqi_clean(df_aqi).groupby(pd.Grouper(freq='1M')).mean()
df2.plot(figsize=(7,20), subplots=True);
```

The above graph shows us that post 2015, Air Quality tends to follow a seasonal pattern, where it tends to peak towards the beginning and end of the year, with dips during the middle months.

```python
df3 = aqi_clean(df_aqi).groupby(pd.Grouper(freq='3M')).mean()[['AQI']]
df3.plot(figsize=(7,3), subplots=True);
```

We see this pattern more prominently when AQI is aggregated over every 3 month period.

###### Relationship Analysis:

```python
plt.subplots(figsize=(14,12))
sns.heatmap(df_aqi.drop(columns=['Xylene']).corr(),annot=True, cbar_kws={'shrink': 0.5});
```

The above correlation heatmap shows us that the pollutants that affect Air Quality most are Particulate Matter (2.5 and 10), Nitrogen Dioxide and Carbon Monoxide.

### Understanding the problem:

Recently, the [World Health Organisation reworked their guidelines](https://www.moneycontrol.com/news/photos/environment/who-issued-new-air-quality-guidelines-take-a-look-7505231.html) for the limitation on emission of pollutants such as PM2.5 and PM10, and they believe that over 80% of the deaths caused due to air pollution could be avoided if the new guidelines are followed.

The proposed guidelines require an annual average of:

* **PM 2.5** - 5 μg/m³ (previously 10)
* **PM 10** - 15 μg/m³ (previously 20)
* **NO2** - 10 μg/m³ (previously 40)

So, how is Bangalore performing on these scales? 
(Spoiler: Not too good.)

```python
df3 = aqi_clean(df_aqi).groupby(pd.Grouper(freq='1Y')).mean()[['PM2.5','PM10','NO2']]
df3.plot(figsize=(12,6), subplots=True);
```

At its best, Bangalore has reached annual averages of ~30 and ~70 in PM 2.5 and PM 10, respectively. These figures are ~6x and ~4.5x above the recommended limit and even with NO2, where it is faring relatively better, it is at least 2x the recommended limit.

The consequences of these breaches of guidelines are dire, too as [last year alone, Bangalore lost 12,000 lives to air pollution](https://www.thehindu.com/news/cities/bangalore/air-pollution-led-to-12000-deaths-in-bengaluru-last-year/article33874517.ece), which is alarming given the fact that 2020 is the year Bangalore fared best in terms of emissions.

### So, what can we do about this?

To explore possible solutions to the air pollution problem, we will use the COVID-19 lockdowns as an experimental set-up with pre-lockdown acting as the control group and post-lockdown acting as the test group.

##### Is there actually a difference between air quality before and after the lockdowns?

```python
lockdown_19 = dt.datetime(2019, 3, 25)
end = dt.datetime(2019,7,1)
lockdown = dt.datetime(2020, 3, 25)

aqi_20 = df_aqi[df_aqi['Date'].dt.year == 2020][['Date','AQI']]
aqi_19 = df_aqi[(df_aqi['Date'] <= end) & (df_aqi['Date'].dt.year == 2019)][['Date','AQI']]

aqi_20['Date'] = aqi_20.Date.apply(lambda x: x.strftime('%m-%d'))
aqi_19['Date'] = aqi_19.Date.apply(lambda x: x.strftime('%m-%d'))

plt.rcParams["figure.figsize"] = (20,8)
plt.plot(aqi_20['Date'], aqi_20['AQI'], label='2020', color='green')
plt.plot(aqi_19['Date'], aqi_19['AQI'], linestyle=":", color='black', label='2019')
plt.xticks(rotation=90)

for label in plt.gca().get_xaxis().get_ticklabels()[::2]:
    label.set_visible(False)
    
plt.axvline('03-25',ls='--', c='indianred', label='Initial Lockdown')
plt.axvline('06-08',ls='--', c='c', label='Inital Unlock')

plt.legend();
```

**Note**: As we discussed earlier, Air Quality in Bangalore tends to follow a seasonal pattern, due to this it wouldn't make much sense to contrast the AQI from March - July to samples from all the months. Thus, the above graph compares Bangalore's AQI from 2020 to the same in 2019.

From January through March, air quality in 2020 behaved in very similar ways when compared to 2019. However, when the initial lockdown hit, the air quality in 2020 improved drastically and was lower than its 2019 counterpart every day until June 11th 2020, which was 2 days after the lockdown was lifted.

This shows a very significant difference in the air quality during the lockdown but this also brings up another important question, is this difference **statistically** significant? Statistical significance is important because it acts as a distinction between outcomes as a consequence of chance, or as a consequence of the factors one is trying to study. To find this out, we will conduct a hypothesis test.

### Hypothesis Testing:

* **Null Hypothesis:** There is no statistically significant difference between the average AQI during the lockdown period in 2020 and the same time period in 2019.

* **Test**: One-way analysis of variance (ANOVA).
* **α**: 0.05

```python
test_19 = df_aqi[(df_aqi['Date'] <= end) & (df_aqi['Date'] >= lockdown_19)][['Date','AQI']]
test_20 = df_aqi[df_aqi['Date'] >= lockdown][['Date','AQI']]

test_19['Group'] = '2019'
test_20['Group'] = '2020'
 
F, p = st.f_oneway(test_19['AQI'], test_20['AQI'])
print('p-value of the ANOVA test:',p)

anova = test_19.append(test_20)
anova.reset_index(drop=True, inplace=True)
anova.groupby('Group').mean().sort_values('AQI', ascending=False).plot(kind='bar', figsize=(10,7));
```

The p value for this test comes out to be  2.7 * 10⁻²⁷, which is lower than our defined value of α (0.05), thus we reject the null hypothesis.

### Breaking down the causal factors:

The hypothesis test shows us that COVID-19 had a statistically significant impact on air pollution/air quality in Bangalore. However, the pandemic and the lockdown are both extreme events, which make them very tough to draw insight toward individual action from.

##### What is nighttime lights data?

As the name suggests, this data is a remote sensing of lights during the night. This light can emerge from various sources, such as moonlight, or light directly emitted from transport and buildings, and reflection of light from the ground. [Night time lights data has been shown to be correlated to socio-economic activity](https://rbidocs.rbi.org.in/rdocs/Content/PDFs/01AR30072019EF4B60BF96E548F284D2C95EB59DD9A9.PDF) and intuitively, it also makes sense that it is reflective of domestic transport since it measures lights in the near infrared range which encompasses the spectrum of most human-made lights.

An example of night-time lights data can be seen below:

```python
map1 = geemap.Map(center=[lat,lon],zoom=8)
map1.add_basemap('SATELLITE')
viirs2020_2 = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterDate("2020-02-01","2020-02-29").select('avg_rad').first()
map1.addLayer(viirs2020_2, {}, "viirs")
map1
```

Specifically, we will be using The Visible Infrared Imaging Radiometer Suite (VIIRS) Day/Night Band, as it provides better spatial and temporal resolution than its counterpart.

##### How has the pandemic impacted night-time lights?

To assess the effect of the pandemic, we will contrast the monthly composite from Feb 2020 (the last full month with no lockdown) to the one from May 2020 (the last full month with lockdown). 

We will also be normalising the composite by subtracting its mean radiance and dividing it by the standard deviation because as we can see above - Bangalore is very radiant and it can be difficult to see trends within the city at such a large scale.

```python
viirs2020_2 = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterDate("2020-02-01","2020-02-29").select('avg_rad').first()
viirs2020_5 = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterDate("2020-05-01","2020-05-31").select('avg_rad').first()

blr = ee.FeatureCollection("users/ujavalgandhi/e2e/ne_10m_urban_areas").filter(ee.Filter.eq('system:index', '00000000000000002bf8')).geometry()
ntl_blr = viirs2020_2.clip(blr)

mu = ntl_blr.reduceRegion(reducer=ee.Reducer.mean())
std = ntl_blr.reduceRegion(reducer=ee.Reducer.stdDev())
mu = ee.Number(mu.get('avg_rad'))
std = ee.Number(std.get('avg_rad'))
```

```python
print(f"The mean of Bangalore's average radiance in Feb 2020 is {round(mu.getInfo(),2)}.")
```

```python
viirs2020_2_mask = viirs2020_2.subtract(mu).divide(std)
viirs2020_5_mask = viirs2020_5.subtract(mu).divide(std)

map2 = geemap.Map(center=[lat,lon],zoom=11)
map2.add_basemap('SATELLITE')

pre_tile = geemap.ee_tile_layer(viirs2020_2_mask, {}, 'DMSP-OLS 1996', opacity=0.6)
post_tile = geemap.ee_tile_layer(viirs2020_5_mask, {}, 'DMSP-OLS 2010', opacity=0.6)

map2.split_map(left_layer=pre_tile, right_layer=post_tile)
map2
```

In just 3 months, Bangalore's change in night-time lights is drastic. The slider can be used to see the full extent of this change. This could be due to number of reasons, ranging from reduced trasport to the outflow of people heading back to their home towns, and so on and so forth. 

```python
viirs = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterDate('2020-02-01','2020-05-31')

first_img = viirs.sort('system:time_end').first()
last_img = viirs.sort('system:time_end',False).first()
viirs_slope = (last_img.select('avg_rad').subtract(first_img.select('avg_rad'))).divide(viirs.size().getInfo())

viirsMap = geemap.Map(center=[lat,lon],zoom=10)
viz_params = {'min':-1,
             'max':1,
             'palette':['1d4877','1b8a5a','f68838','ee3e32']}

colors = viz_params['palette']
vmin = viz_params['min']
vmax = viz_params['max']

viirsMap.addLayer(viirs_slope, viz_params, 'Feb-May VIIRS-DNB monthly rate of change',opacity=.75)
viirsMap.addLayerControl()
viirsMap.add_colorbar_branca(colors=colors, vmin=vmin, vmax=vmax, layer_name="Feb-May VIIRS-DNB monthly rate of change")
viirsMap
```

Between Feb and May, we see that radiance reduced significantly in the urban areas of Bangalore.

The above maps indicate that during the lockdown there was a significant reduction in the quantity and spread of night-time lights, which for the purposes of this study act as a representative of economic activity and domestic transport. However, this relationship between air quality and night-time lights is still seen as the consequence of the extreme event of the lockdown. To look at the relationship between the intensity of change in air quality with the intensity of change in night-time lights, we will look at the correlation between these two variables.

##### Can we quantify night-time lights?

Yes, we can use the reduceRegion function provied by Google's Earth Engine to to reduce a given region (Bangalore urban) to its mean radiance. This can further be used to measure quantitative relationships with air quality and particulate matter ejections.

```python
viirs = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterDate('2015-03-01','2020-7-31')

def poi_mean(img):
    mean = img.reduceRegion(reducer=ee.Reducer.mean(), geometry=blr).get('avg_rad')
    return img.set('date', img.date().format()).set('mean',mean)

poi_reduced_imgs = viirs.map(poi_mean)
nested_list = poi_reduced_imgs.reduceColumns(ee.Reducer.toList(2), ['date','mean']).values().get(0)
df_nl = pd.DataFrame(nested_list.getInfo(), columns=['date','mean'])
df_corr = pd.DataFrame({'ntl':list(df_nl['mean']), 'AQI':df2.AQI, 'PM2.5':df2['PM2.5'], 'PM10':df2['PM10']})
# df_corr = df_corr[df_corr.index.year >= 2019]
df_corr.plot(subplots=True, figsize=(10,7));
```

##### How do night time lights affect air quality parameters?

The above graph seems to confirm our suspicion that even with small changes in night-time lights, we can still see changes in air quality, and more specifically with particulate matter levels. Next, let's look at the data at a more granular level.

```python
g = sns.PairGrid(df_corr, aspect=1.2)
g.map(sns.scatterplot);
```

When looked at at a more granular level, the data indicates that not only do small changes have an impact, but rather that small changes can impact particulate matter levels more efficiently than massive ones. 

Based on data from 2015 - 2020, it is observed that taking the mean radiance in Bangalore from >50 to ~40 has almost as much impact on PM2.5 and PM10 levels as it reaching its lockdown lows of <10. This tells us that these variables are non-linearly correlated, so we will try to quantify this correlation using Spearman's Rank Correlation rather than Pearson's.

```python
plt.subplots(figsize=(7,5))
sns.heatmap(df_corr[df_corr.index.year >= 2020].corr(method='spearman'), annot = True, center=0, cbar_kws={'shrink':0.5});
```

In 2020, the average radiance displays a 100% rank correlation with PM2.5, and 96% rank correlation with PM10 and AQI.

This also opens avenues that can be researched by corporations for better ESG implementation through ideas such as an increase in the number of remote jobs to reduce mandatory travel, or reducing the amount of urban vegetation depleted caused due to industry expansion.


###### What's vegetation got to do with this?

Vegetation and night-time lights are correlated since vegetation is less reflective than urbanized land. Further, there have been reports that suggest that [industrial expansion and urbanization leads to loss of photosynthetic resources](https://storymaps.arcgis.com/stories/733157be098144b7af2a591460bb6c00) and that [vegetation (when not allowed to decay) can help with carbon sequestration](https://climate.nasa.gov/news/3043/study-urban-greenery-plays-a-surprising-role-in-greenhouse-gas-emissions/) making it important to air quality as well.

Analyzing vegetation during the pandemic (a period where economic activity came to a standstill) can thus represent a potentially massive avenue for tackling the problem of air pollution by participation in reforestation programs by governments and NGOs. For this task, we will be using rasterio and Sentinel-2 satellite imagery sourced from [Sentinel Hub's EO Browser](https://apps.sentinel-hub.com/eo-browser/) to calculate a dimensionless metric known as Normalized Difference Vegetation Index
(or NDVI).

```python
import rasterio as rio
import rasterio.plot as rplt
import numpy as np
import geopandas as gpd
```

##### What is NDVI?

NDVI is a dimensionless metric that uses the red band and the near infrared band from multispectral satellite images to estimate the green cover of the respective EO data. This works due to the less-reflective nature of vegetation. To start of, we will procure the red (4th) and NIR (8th) bands from the Sentinel-2 satellites for Bangalore.

**Note**: For this section, I have decided to use images from February from 2018-2020 because:
1. As discussed earlier, it is the last full month with no state-enforced lockdown in 2020.
2. Due to the local climate of Bangalore, the months of May-October are extremely cloudy and that can skew the image of vegetation.
3. Since vegetation is a long-term process, i.e, the changes are gradual, we will be using one composite per year to measure these gradual changes.

More information on NDVI can be found [here](https://www.sciencedirect.com/topics/earth-and-planetary-sciences/normalized-difference-vegetation-index).

```python
def get_bands(year):
    filepath = f'samples/Feb{year}'
    try:
        for file in os.listdir(filepath): 
            if 'B04' in file:
                img = rio.open(f'{filepath}/{file}')
                red = img.read(1).astype('f4')
            elif 'B08' in file:
                img = rio.open(f'{filepath}/{file}')
                nir = img.read(1).astype('f4')
    except FileNotFoundError:
        print('Directory does not exist, try a different directory.')
        
    return red, nir

red, nir = get_bands(2020)

fig = plt.figure(figsize=(15,7.5))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(red, cmap='pink')
ax1.axis('off')
ax1.set_title('Bangalore - Red Band')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(nir, cmap='pink')
ax2.axis('off')
ax2.set_title('Bangalore - NIR Band');
```

The above graph displays the contrast between the red and NIR bands, to calculate the NDVI we will subtract the pixels of these two bands and then divide the difference by their sum, giving us the formula:
##### **NDVI = (NIR - Red) / (NIR + Red)**

```python
red2, nir2 = get_bands(2021)

ndvi = (nir - red)/(nir + red)
ndvi2 = (nir2 - red2)/(nir2 + red2)

fig = plt.figure(figsize=(15,7))
ax1 = fig.add_subplot(1,3,1)
ax1.imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)
ax1.axis('off')
ax1.set_title("NDVI - Feb 2020")
ax2 = fig.add_subplot(1,3,2)
ax2.imshow(ndvi2, cmap='RdYlGn', vmin=-1, vmax=1)
ax2.axis('off')
ax2.set_title("NDVI - Feb 2021")
ax3 = fig.add_subplot(1,3,3)
ax3.imshow(ndvi2-ndvi, cmap='bwr', vmin=-1, vmax=1)
ax3.axis('off')
ax3.set_title("NDVI Difference");
```

In Bangalore, some instances of strong deforestation (represented by dark blue) around the east ends of the city (i.e, the industrialised complex of Whitefield) are observed, however these instances are rare and in contrast, it is seen that the overall landscape of Bangalore has gained many (less dense) patches of vegetation.

```python
ndvi_diff = ndvi2-ndvi
bangalore = gpd.read_file('data/BBMP_Wards_2011_region.json')


fig, ax = plt.subplots(figsize=(10,10))
bangalore.geometry.boundary.plot(color=None,edgecolor='k',linewidth = 0.5,ax=ax)
plt.sca(ax)
plt.imshow(ndvi_diff, cmap='bwr', extent=(77.45, 77.79, 12.83, 13.14), vmin=-1, vmax=1)
plt.colorbar(shrink=0.5)
plt.axis('off')
plt.title("Difference in NDVI from Feb 2020 to Feb 2021");
```

The above overlay of the wards of urban Bengaluru is sourced from [here](https://github.com/openbangalore/bangalore/blob/master/bangalore/GIS/BBMP_Wards_2011_region.json). It helps provide context as to which wards/areas see what kind of trend in vegetation during post-pandemic life.

For more context we will quanitfy the NDVI in the same way that we quanitified night-time lights.

##### How did NDVI change during the pandemic?

```python
def ndvi_mean(year):
    red, nir = get_bands(year)
    ndvi = (nir - red)/(nir + red)
    return np.nan_to_num(ndvi, nan=0).mean()

ndvi_years = list()
ndvi_means = list()

for i in range(2018,2022):
    ndvi_years.append(i)
    ndvi_means.append(ndvi_mean(i))

ndvi_df = pd.DataFrame({'Year':ndvi_years, 'Avg_ndvi':ndvi_means})
ndvi_df.plot(x='Year', y='Avg_ndvi', figsize=(12,6))
plt.title('Change in NDVI from Feb 2018 to Feb 2021');
```

While in most years, the difference between the average of NDVI values has been between 1-2%, during the pandemic the vegetation index saw a rise of ~6%.

**How did the NDVI affect air quality in these years?**

```python
df_years = df2[df2.index.month == 2][['AQI', 'NO2']]
df_years = df_years[df_years.index.year > 2017]
df_years.reset_index(inplace=True)
df_years['ndvi'] = ndvi_df['Avg_ndvi'].loc[0:2]
df_years.loc[len(df_years.index)] = [dt.date(2021,2,28), 71.43, 26.11, ndvi_df['Avg_ndvi'].loc[3]] # values taken and averaged manually from https://cpcb.nic.in/real-time-air-qulity-data/
df_years.set_index('Date').plot(figsize=(12,7), subplots=True)
plt.title('Change in Monthly Average AQI, NO2 and NDVI for the month of February from 2018 to 2021.', y = 3.55);
```

The above graph displays a **strong** inverse linear relationship between air quality and nitrogen dioxide, and average NDVI. To quantify this relationship, we will use Pearson's correlation coefficient.

```python
plt.subplots(figsize=(7,6))
sns.heatmap(df_years.corr(), annot=True, cbar_kws={'shrink':0.5});
```

Despite the obvious caveats of a small sample size and somewhat unreliable data for 2021, the findings here are at the very least worth investigating as the correlation heatmap displays a correlation of -94% between NDVI and AQI, and -97% between NDVI and NO2.

### Conclusion:

* The WHO has set new guidelines for annual emissions of pollutants: **PM 2.5 - 5μg/m³, PM 10 - 15μg/m³, NO2 - 10μg/m³.**


* **Bangalore has struggled** to adhere to these guidelines from 2015 - 2020. However, massive strides in the right direction were observed during the COVID lockdowns. Despite this, last year alone, **12,000 lives were lost** to air pollution.


* **Hypothesis testing** was conducted to deduce whether there was a statistically significant change in **air pollution during the lockdowns**. The null hypothesis was rejected with a **p-value of  2.7 * 10⁻²⁷**.


* When looking at **satellite imagery** in the form of the **VIIRS-DNB night-light data** (which offers insight into economic activity, population trends and domestic transportation), these bands showed **prominent rank correlation** with **particulate matter ejections** and it stands to reason that even with **smaller (but widespread) changes** in the way we conduct economic activity, we could **help reduce the emissions** of these particles to get closer to the recommended limit. **Some suggestions** include but are not limited to:
    * Modifying our means of transport, including carpooling and switching to cycles/electric vehicles. 
    * Increasing the ratio of remote jobs to curb the explosive population growth of Bangalore due to an influx of working professionals (while also curbing mandatory travel).
    * Working on restoring the large water-bodies of Bangalore (since Water Bodies reflect less light, they also correlate with the ratio of night-time lights).


* **Urban vegetation** in Bangalore was also analysed and **very strong inverse linear correlation** was observed between the mean **NDVI** of a given raster and the amount of **NO2** emissions and also the **Air Quality** Index. Despite the analysis being conducted on a small sample size, the strength of this correlation is a promising avenue to explore in more depth, as it **strongly incentivizes:**
    * Preserving hyperlocal vegetation such as parks and gardens. This also carries with it the cultural value of Bangalore's image as the 'Garden City'.
    * Supporting NGOs in stopping large scale deforestation projects.
    * Supporting other corporations that have reforestation programs.
    * Cultivating personal plant-life, since any step (done by enough people) pushees things in the right direction.


* This indicates that there is a strong possibility for improvement in particulate matter emissions by leveraging better resource management in urbanization and transport, and in nitrogen dioxide levels by improving the quality and quantity of urban vegetation.
